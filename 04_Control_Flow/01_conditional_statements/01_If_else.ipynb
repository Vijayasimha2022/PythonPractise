# Python Control Flow: `if` and `else`
## Overview

In Python, conditional statements direct the flow of your program. Unlike C-family languages, Python drops parentheses around conditions and relies entirely on a colon (`:`) and **strict indentation** to define code blocks instead of curly braces `{}`.

Additionally, Python condenses `else if` into the single keyword **`elif`**.

---

## 1. Syntax & Structure

The `if-elif-else` ladder evaluates conditions top-to-bottom. As soon as one condition evaluates to `True`, its corresponding block executes, and the rest of the ladder is entirely bypassed.

```python
status_code = 403

if status_code == 200:
    # 4 spaces (standard indentation)
    print("Success: Data retrieved.")
elif status_code == 401:
    print("Error: Unauthorized.")
elif status_code == 403:
    print("Error: Forbidden access.")
else:
    print("Error: Unknown status.")

```

### Key Engineering Rules:

1. **The Colon (`:`):** Every `if`, `elif`, and `else` statement must end with a colon. Forgetting this will raise a `SyntaxError`.
2. **Indentation:** All code within the block must be indented consistently (4 spaces is the PEP 8 standard).
3. **Optional Components:** You can have an `if` statement without an `elif` or an `else`. You can also have multiple `elif` blocks.

---

## 2. "Truthy" and "Falsy" Evaluation (Crucial Concept)

Python does not require conditions to strictly evaluate to a boolean `True` or `False`. Instead, it evaluates the "Truthiness" of an object. This allows for highly concise backend checks (like verifying if a list or string is empty).

### Falsy Values

If you pass these into an `if` statement, they evaluate to `False`:

* Numeric zero: `0`, `0.0`
* Empty sequences/collections: `""` (empty string), `[]` (empty list), `{}` (empty dict)
* The `None` type (Python's equivalent of null)
* `False`

### Truthy Values

Virtually everything else evaluates to `True`.

```python
# Instead of checking lengths or nulls explicitly:
user_roles = []
username = "admin_user"

# Because user_roles is an empty list, it evaluates to False
if not user_roles:
    print("No roles assigned.")

# Because username is a non-empty string, it evaluates to True
if username:
    print("Username is valid.")

```

---

## 3. Inline Conditionals (The Python Ternary Operator)

When you need to assign a variable based on a simple condition, using a full `if-else` block is overly verbose. Python supports conditional expressions in a single line.

**Syntax:** `value_if_true if condition else value_if_false`

```python
latency_ms = 120

# Python's equivalent of (condition) ? true_val : false_val
health_status = "Healthy" if latency_ms < 200 else "Degraded"

print(health_status) # Output: Healthy

```

---

## 4. Logical Operators within `if`

To combine multiple conditions, Python uses plain English keywords instead of symbols like `&&`, `||`, or `!`.

* **`and`**: True if both sides are True.
* **`or`**: True if at least one side is True.
* **`not`**: Reverses the boolean value.

```python
is_authenticated = True
permissions = ["read", "write"]

if is_authenticated and "delete" not in permissions:
    print("Action denied: Insufficient privileges.")

```


# Python Control Flow: Nested `if` and `elif`

## Overview

A **Nested Conditional** is simply an `if`, `elif`, or `else` statement placed inside another `if`, `elif`, or `else` block. You use nesting when a secondary condition should only be checked *after* a primary condition has been met.

While powerful, nesting requires careful attention to indentation, as Python uses whitespace to determine which block a statement belongs to.

---

## 1. Syntax & Indentation

Every time you nest a conditional, you must increase the indentation level (standard practice is an additional 4 spaces).

```python
# Level 1 (Outer)
if primary_condition:
    
    # Level 2 (Inner)
    if secondary_condition:
        print("Both conditions met.")
    elif alternative_secondary:
        print("Primary met, alternative secondary met.")
    else:
        print("Only primary met.")

else:
    print("Primary condition failed. Inner conditions are never checked.")

```

---

## 2. Engineering Example: Dependent Logic

Nesting is highly useful for hierarchical checks, such as verifying a user's login status before checking their specific access tier.

```python
is_logged_in = True
subscription_tier = "Pro"

# Primary Check
if is_logged_in:
    print("Welcome back, user!")
    
    # Secondary Check (Dependent on being logged in)
    if subscription_tier == "Free":
        print("Showing ads.")
    elif subscription_tier == "Pro":
        print("Ad-free experience enabled. Accessing premium features.")
    else:
        print("Please verify your subscription.")
        
else:
    print("Please log in to continue.")

```

---

## 3. The "Arrow Anti-Pattern" (Engineering Best Practice)

While nesting is syntactically valid, deep nesting (more than 2 or 3 levels) is considered bad practice. It creates the "Arrow Anti-Pattern," where your code visually points to the right like an arrow `>`. This makes the logic incredibly difficult to read, test, and debug.

### The Problem (Deep Nesting)

```python
if file_exists:
    if is_readable:
        if file_size < max_size:
            # Do work
            print("Processing file...")

```

### The Fix: Flattening via Logical Operators

If the inner conditions are strictly dependent on each other, combine them using the `and` operator to keep the code flat.

```python
if file_exists and is_readable and file_size < max_size:
    # Do work
    print("Processing file...")

```

### The Fix: Flattening via Guard Clauses (Early Returns)

In functions, it is standard practice to check for the *failure* conditions first and exit immediately, leaving the main logic un-indented at the bottom.
