# Python Arithmetic Operators

Python provides a rich set of arithmetic operators to handle numerical computations. While operators like addition and subtraction are universal across programming languages, Python introduces specific operators for precise division control and exponentiation that distinguish it from languages like C++ or Java.

---

## 1. The Standard Operators

These perform exactly as you would expect from basic mathematics.

* **Addition (`+`):** Adds two operands.
* **Subtraction (`-`):** Subtracts the right operand from the left.
* **Multiplication (`*`):** Multiplies two operands.

### Engineering Syntax Example

```python
# Terminology: 'a' and 'b' are Operands. '+' is the Operator.
a = 10
b = 15

# The entire line is an Expression
c = a + b  # c becomes 25

```

---

## 2. The Division Triad (Crucial Distinction)

Python 3 makes a strict distinction between "True Division" (Float) and "Floor Division" (Integer). Understanding the difference between these three operators is essential for avoiding bugs in algorithms that require whole numbers (like array indexing).

### A. True Division (`/`)

Always returns a **Float**, even if the numbers divide evenly.

```python
x = 14
y = 4

result = x / y
print(result) # Output: 3.5

# Even if it divides perfectly:
print(8 / 2)  # Output: 4.0 (Notice the .0)

```

### B. Floor (Integer) Division (`//`)

Returns the **quotient** rounded *down* to the nearest whole integer. It strips away the decimal portion.

```python
x = 14
y = 4

result = x // y
print(result) # Output: 3

```

### C. Modulus (`%`)

Returns the **remainder** of a floor division.
*Engineering Note:* This is one of the most frequently used operators in programming, essential for determining if a number is even/odd (`x % 2 == 0`) or for keeping values within a specific cyclical range (like a clock).

```python
x = 14
y = 4

# 4 goes into 14 three times (12), leaving a remainder of 2
result = x % y
print(result) # Output: 2

```

---

## 3. Exponentiation (`**`)

Unlike many languages that require importing a `math` module to perform exponents (e.g., `math.pow()`), Python has a built-in operator for it.

* **Syntax:** `base ** exponent`

```python
# 2 to the power of 5 (2 * 2 * 2 * 2 * 2)
result = 2 ** 5
print(result) # Output: 32

# It also works for finding roots (e.g., square root is power of 0.5)
square_root = 25 ** 0.5
print(square_root) # Output: 5.0

```

| Operator | Name | Example (`a=14`, `b=4`) | Result |

| `*` | Multiplication | `a * b` | `56` |

| `//` | Floor Division | `a // b` | `3` (Integer Quotient) |
