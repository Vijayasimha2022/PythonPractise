# Python Expressions, Precedence, and Associativity

An **Expression** is any combination of operators (like `+`, `*`) and operands (like variables or literal numbers) that evaluates to a single value.

When an expression contains multiple operators, Python needs a strict set of rules to determine the order of evaluation so that the calculation is predictable and consistent. These rules are defined by **Operator Precedence** and **Associativity**.

---

## 1. Operator Precedence (Order of Operations)

Precedence defines which operators bind tighter to their operands. It is the programming equivalent of the mathematical PEMDAS/BODMAS rules. Operators with higher precedence are evaluated before operators with lower precedence.

### The Arithmetic Precedence Hierarchy (Highest to Lowest)

1. **Parentheses:** `()`
2. **Exponentiation:** `**`
3. **Multiplication, Division, Floor Division, Modulus:** `*`, `/`, `//`, `%`
4. **Addition, Subtraction:** `+`, `-`

### Engineering Example

```python
# The expression:
result = 2 + 3 * 5

# How Python evaluates it:
# 1. Multiplication has higher precedence than Addition.
# 2. Python calculates (3 * 5) = 15
# 3. Python calculates 2 + 15 = 17

print(result) # Output: 17

```


## 2. Associativity (Tie-Breakers)

What happens if an expression has multiple operators with the *exact same* precedence?
For example: `10 - 4 + 2` or `2 ** 3 ** 2`.

This is where **Associativity** comes in. It determines the direction in which operators of the same precedence are evaluated.

### A. Left-to-Right Associativity (The Default)

Almost all arithmetic operators (`+`, `-`, `*`, `/`, `//`, `%`) evaluate from left to right.

```python
# The expression:
result = 10 - 4 + 2

# Both '-' and '+' have the exact same precedence.
# Because associativity is Left-to-Right, Python evaluates the left-most operator first.
# 1. 10 - 4 = 6
# 2. 6 + 2 = 8

print(result) # Output: 8

```

### B. Right-to-Left Associativity (The Exception)

The **Exponentiation (`**`)** operator is unique: it evaluates from **Right to Left**.
This aligns with standard mathematical notation for stacked exponents ($x^{y^z}$).

```python
# The expression:
result = 2 ** 3 ** 2

# Because '**' associates Right-to-Left, Python evaluates the right-most operator first!
# 1. Python calculates 3 ** 2 = 9
# 2. Python calculates 2 ** 9 = 512

print(result) # Output: 512

# This is completely different from evaluating left-to-right:
# (2 ** 3) ** 2  =>  8 ** 2 => 64 (Incorrect mathematically)

```

*Note: Assignment operators (like `=`, `+=`) also associate right-to-left (e.g., `a = b = c = 5` evaluates `c=5`, then `b=c`, then `a=b`).*



## 3. The Power of Parentheses `()`

Parentheses have the highest precedence in Python. They are the ultimate override tool.

If you need an addition operation to happen before a multiplication operation, you must wrap the addition in parentheses.

### Engineering Best Practice

Even if you have memorized the precedence table perfectly, **use parentheses to make your intent explicit.** Code is read far more often than it is written. Relying on obscure precedence rules makes code harder for other engineers to read and maintain.

```python
# BAD: Relies on implicit precedence knowledge. Hard to read quickly.
# Is it (5+2) * (3-8) / 2? Or 5 + (2*3) - (8/2)?
x = 5 + 2 * 3 - 8 / 2

# GOOD: Explicitly parenthesized. Intent is obvious.
x = 5 + (2 * 3) - (8 / 2)

```

### Forcing Precedence

```python
# Default evaluation (Multiplication first)
print(5 + 2 * 3) # Output: 11

# Forced evaluation (Addition first)
print((5 + 2) * 3) # Output: 21

```
