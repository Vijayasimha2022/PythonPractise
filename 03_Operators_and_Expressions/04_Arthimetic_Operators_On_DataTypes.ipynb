# String Operations and Data Type Interactions in Python

In this module, we explore how Python's standard arithmetic operators behave when applied to non-numerical data types, specifically Strings. We also examine how Python handles mathematical expressions that mix different data types (Type Coercion).

---

## 1. Arithmetic Operators on Strings

Python uses a concept called **Operator Overloading**. This means that the exact same symbol (like `+`) does entirely different things depending on the data types it is applied to.

If you use `+` on two integers, it performs mathematical addition. If you use `+` on two strings, it performs **Concatenation**.

### A. String Concatenation (`+`)

Concatenation simply means joining strings end-to-end.

**Rule:** You can *only* concatenate a string with another string. Attempting to add an integer to a string will result in a `TypeError`.

```python
s1 = "Hello"
s2 = " World"

# Valid Concatenation
result = s1 + s2 
print(result) # Output: "Hello World"

# Invalid: Adding int to string
# s1 + 10  -> Raises TypeError

# The Fix: Type Conversion
result = s1 + str(10)
print(result) # Output: "Hello10"

```

### B. String Repetition (`*`)

You can use the multiplication operator to repeat a string a specific number of times.

**Rule:** You can *only* multiply a string by an Integer. Attempting to multiply a string by a float (e.g., `5.5`) will result in a `TypeError`.

```python
s = "A"

# Valid Repetition
print(s * 5) # Output: "AAAAA"

# Invalid: Multiplying by Float
# print(s * 5.5) -> Raises TypeError

```


## 2. Arithmetic on Other Data Types

### A. Floats (`//` and `%`)

While standard division (`/`) on floats is straightforward, applying Floor Division (`//`) or Modulus (`%`) to floats behaves identically to integers, but the returned result is still a float.

```python
# Floor Division: How many full times does 3.1 go into 10.5? (3 times)
result_floor = 10.5 // 3.1
print(result_floor) # Output: 3.0 (Note it remains a float)

# Modulus: What is the remainder? (10.5 - (3.1 * 3) = 1.2)
result_mod = 10.5 % 3.1
print(result_mod) # Output: 1.2

```

### B. Booleans

In Python, the boolean values `True` and `False` are technically just disguised integers.

* `True` evaluates to `1`
* `False` evaluates to `0`

You can perform standard arithmetic on them, and the result will be an Integer.

```python
print(True + True)   # Output: 2 (1 + 1)
print(True + 5)      # Output: 6 (1 + 5)
print(False * 100)   # Output: 0 (0 * 100)

```

### C. Complex Numbers

You can add, subtract, multiply, and divide complex numbers natively.
**Exception:** You *cannot* use Floor Division (`//`) or Modulus (`%`) on complex numbers.

```python
c1 = 2 + 3j
c2 = 1 + 1j

print(c1 + c2) # Output: (3+4j)

```

## 3. Resultant Data Types (Type Coercion/Promotion)

When an expression contains mixed data types, Python automatically "promotes" the result to the most complex (or "largest") data type present in the expression to prevent data loss.

### The Type Hierarchy (Smallest to Largest)

1. **Boolean** (Smallest)
2. **Integer**
3. **Float**
4. **Complex** (Largest)

### Engineering Examples

```python
# Example 1: Boolean + Integer
# Hierarchy: Int > Bool. Result is Integer.
result = True + 5
print(type(result)) # <class 'int'>

# Example 2: Integer + Float
# Hierarchy: Float > Int. Result is Float.
result = 3 + 2.5
print(type(result)) # <class 'float'>

# Example 3: Float + Complex
# Hierarchy: Complex > Float. Result is Complex.
result = 2.5 + (1 + 2j)
print(type(result)) # <class 'complex'>

```

*(Note: As mentioned previously, True Division `/` will ALWAYS result in a float, regardless of the inputs).*
